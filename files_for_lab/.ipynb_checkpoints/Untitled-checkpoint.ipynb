```python
def differentiate_numerical_types(df):
    # Select numerical columns of type float64
    continuous_df = df.select_dtypes(include=['float64'])
    
    # Select numerical columns of type int64
    discrete_df = df.select_dtypes(include=['int64'])
    
    return continuous_df, discrete_df

# Call the function to get continuous and discrete variables
continuous_df, discrete_df = differentiate_numerical_types(customer_df)

# Now, continuous_df contains float64 columns, and discrete_df contains int64 columns
```

```python
def plot_discrete_variables(df):
    # Iterate through columns in the DataFrame
    for column in discrete_df.columns:
        # Check if the column is discrete (e.g., int or categorical)
        if discrete_df[column].dtype in [int, 'int64', 'int32']:
            # Create a count plot (bar plot) for the discrete variable
            plt.figure(figsize=(8, 6))
            sns.histplot(data=df, x=column)
            plt.title(f'Distribution of {column}')
            plt.xlabel(column)
            plt.ylabel('Count')
            plt.xticks(rotation=45)
            plt.show()

# Call the function to plot discrete variables in the "discrete_df" DataFrame
plot_discrete_variables(discrete_df)
```

```python
#Create a function to plot every discrete variable. Do the same with the continuous variables 

def plot_discrete_variables_boxplot(df):
    # Iterate through columns in the DataFrame
    for column in df.columns:
        # Check if the column is discrete (e.g., int or categorical)
        if df[column].dtype in [int, 'int64', 'int32']:
            # Create a box plot for the discrete variable
            plt.figure(figsize=(8, 6))
            sns.boxplot(data=df, x=column)
            plt.title(f'Box Plot of {column}')
            plt.xlabel(column)
            plt.show()

# Call the function to plot discrete variables using box plots in the "discrete_df" DataFrame
plot_discrete_variables_boxplot(discrete_df)
```

```python
def plot_continuous_variables(df):
    # Iterate through columns in the DataFrame
    for column in df.columns:
        # Check if the column is continuous (e.g., float)
        if df[column].dtype in [float, 'float64', 'float32']:
            # Create a distribution plot (histogram) for the continuous variable
            plt.figure(figsize=(8, 6))
            sns.histplot(data=df, x=column, kde=True)
            plt.title(f'Distribution of {column}')
            plt.xlabel(column)
            plt.ylabel('Density')
            plt.show()

# Call the function to plot continuous variables in the "continuous_df" DataFrame
plot_continuous_variables(continuous_df)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

def plot_continuous_variables_boxplot(df):
    # Iterate through columns in the DataFrame
    for column in df.columns:
        # Check if the column is continuous (e.g., float)
        if df[column].dtype in [float, 'float64', 'float32']:
            # Create a box plot for the continuous variable
            plt.figure(figsize=(8, 6))
            sns.boxplot(data=df, x=column)
            plt.title(f'Box Plot of {column}')
            plt.xlabel(column)
            plt.show()

# Call the function to plot continuous variables using box plots in the "continuous_df" DataFrame
plot_continuous_variables_boxplot(continuous_df)

```

```python
def find_outliers_iqr(df, factor=1.5):
    # Calculating the first quartile (Q1) and third quartile (Q3) for each column
    Q1 = df.quantile(0.25)
    Q3 = df.quantile(0.75)
    
    # Calculating the IQR (Interquartile Range) for each column
    IQR = Q3 - Q1
    
    # Defining the lower and upper bounds for identifying outliers
    lower_bound = Q1 - factor * IQR
    upper_bound = Q3 + factor * IQR
    
    # Finding outliers by comparing values to the bounds
    outliers = ((df < lower_bound) | (df > upper_bound))
    
    # Count the number of outliers in each column
    outlier_count = outliers.sum()
    
    outlier_info = pd.DataFrame({'Variable': outlier_count.index, 'Outlier Count': outlier_count.values})
    
    return outliers, outlier_info

# Set the IQR factor for outlier detection (default is 1.5)
iqr_factor = 1.5

# Find outliers in the continuous_df DataFrame and get outlier counts
outliers, outlier_info = find_outliers_iqr(continuous_df, factor=iqr_factor)

# Display the DataFrame of outliers (True indicates an outlier)
print("Outliers:")
print(outliers)

# Display the DataFrame with outlier counts
print("\nOutlier Counts:")
outlier_info
```
