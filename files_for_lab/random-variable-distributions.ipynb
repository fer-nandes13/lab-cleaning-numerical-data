```python
1#Use the numerical variables from our dataset this week.

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

numerical_df = pd.read_csv('numerical.csv')
```

```python
numerical_df
```

```python
#2. Check using a distribution plot if the variables fit the theoretical normal or exponential distribution.

#iteration code to check if the columns in numerical_df have a normal or exponential distribution

for column in numerical_df.columns:
    plt.figure(figsize=(8, 4))
    sns.histplot(numerical_df[column], kde=True)
    plt.title(f'Distribution of {column}')
    plt.xlabel(column)
    plt.ylabel('Frequency')
    plt.show()
```
